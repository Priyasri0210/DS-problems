Matplotlib is useful for making plots using the data points

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
z = np.cos(x)
print(y)
print(z)
```

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title, x-axis and t-axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine values')
plt.title('sine wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,'r+')
```

```python
plt.plot(x,y,'g.')
```

```python
#adding multiple plots in same graph
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x),'g--')
plt.plot(x, np.cos(x),'r+')
plt.show()
```

Bar charts

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','Tamil','French','Latin','German']
people = [100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('No. of People')
plt.show()
```

Pie Chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','Tamil','French','Latin','German']
people = [100,50,150,40,70]
ax.pie(people, labels = languages, autopct = '%1.1f%%')
plt.show()
```

Scatter plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
```
