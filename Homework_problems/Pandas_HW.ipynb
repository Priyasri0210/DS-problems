```python
students_df['Average'] = (students_df['MathScore'] + students_df['ScienceScore'] + students_df['EnglishScore'])/3
students_df
```

```python
students_df.info()
```

```python
students_df.sort_values(by='MathScore',ascending=False)
```

```python
students_df.loc[(students_df['City'] == 'Mumbai') & (students_df['ScienceScore'] > 80)]
```

```python
students_df.loc[(students_df['Gender'] == 'Female') & (students_df['Average'] > 85)]
```

```python
students_df['Grade'] = np.where(students_df['Average'] >= 90, 'A', np.where((students_df['Average'] >= 80) & (students_df['Average'] < 90),'B',
                       np.where((students_df['Average'] >=70) &(students_df['Average'] < 80),'C','D')))
students_df
```

```python
students_df.groupby('City')['Average'].mean()
students_df.groupby('City')[['MathScore','ScienceScore','EnglishScore']].mean()
```

```python
students_df[students_df['ScienceScore'] == students_df['ScienceScore'].max()]
```

```python
students_df = students_df.drop('StudentID',axis=1)
students_df
```

```python
students_df['PassesMath']= students_df['MathScore'].apply(lambda x: 'Yes' if x >= 75 else 'No')
students_df
```

```python
students_df.to_csv('Students_processed.csv')
```
